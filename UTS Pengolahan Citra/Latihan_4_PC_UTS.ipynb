```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage.util import random_noise
```

```python
img = data.coins()
img = img.astype(np.uint8)

plt.figure(figsize=(6,5))
plt.imshow(img, cmap='gray')
plt.title("Original - Coins")
plt.axis('off')
plt.show()

```

```python
sigma = 15
noise = np.random.normal(0, sigma, img.shape)
noisy = img + noise
noisy = np.clip(noisy, 0, 255).astype(np.uint8)

plt.figure(figsize=(6,5))
plt.imshow(noisy, cmap='gray')
plt.title("Noisy (Gaussian =15)")
plt.axis('off')
plt.show()
```

```python
# Median filter
median = cv2.medianBlur(noisy, 5)

# Gaussian filter
gaussian = cv2.GaussianBlur(noisy, (5,5), 1.0)

# Bilateral filter
bilateral = cv2.bilateralFilter(noisy, d=9, sigmaColor=75, sigmaSpace=75)
```

```python
edges_noisy = cv2.Canny(noisy, 80, 160)
edges_median = cv2.Canny(median, 80, 160)
edges_gaussian = cv2.Canny(gaussian, 80, 160)
edges_bilateral = cv2.Canny(bilateral, 80, 160)
```

```python
plt.figure(figsize=(15,12))

plt.subplot(3,3,1)
plt.imshow(noisy, cmap='gray')
plt.title("Noisy (=15)")
plt.axis('off')

plt.subplot(3,3,2)
plt.imshow(median, cmap='gray')
plt.title("Median Filter")
plt.axis('off')

plt.subplot(3,3,3)
plt.imshow(gaussian, cmap='gray')
plt.title("Gaussian Filter")
plt.axis('off')

plt.subplot(3,3,4)
plt.imshow(bilateral, cmap='gray')
plt.title("Bilateral Filter")
plt.axis('off')

plt.subplot(3,3,5)
plt.imshow(edges_noisy, cmap='gray')
plt.title("Canny - Noisy")
plt.axis('off')

plt.subplot(3,3,6)
plt.imshow(edges_median, cmap='gray')
plt.title("Canny - Median")
plt.axis('off')

plt.subplot(3,3,7)
plt.imshow(edges_gaussian, cmap='gray')
plt.title("Canny - Gaussian")
plt.axis('off')

plt.subplot(3,3,8)
plt.imshow(edges_bilateral, cmap='gray')
plt.title("Canny - Bilateral")
plt.axis('off')

plt.tight_layout()
plt.show()
```
